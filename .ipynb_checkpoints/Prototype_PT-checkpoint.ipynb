# Pairs Trading

```python
sp_data = pd.read_csv('SP500_Close_Data.csv', index_col = 'Date')
```

## Pairs Trading Overview

So the basic idea in pairs trading is to model the spread of two stocks, i.e $Y_t = \beta X_t$. In what follows $X_t$ will be the **first of the pair alphabetically**. As the pairs are cointegrated, we know that $Y_t - \beta X_t$ is mean reverting. Therefore we will enter our position if $|U_t| > \alpha_\text{E} \sigma_{U}$ and leave once $|U_t| < \alpha_\text{L} \sigma_{U}$. There are two things that need to be calculated:

- $\beta$: for which we will use SKLearn on historical data.
- $\sigma_U$: for which we will use historical data once we have $\beta$. 

For both of these we will use the historical data. Note that we could also caluclate these based on a rolling window. 

Depdening on the sign of $U_t = Y_t - \beta X_t$, will give us our position

- $U_t > 0$ : Short $Y_t$; Long $X_t$.
- $U_t < 0$ : Long $Y_t$; Short $X_t$.

To illustrate this we will plot $U_t$ for a cointegrated pair. 

```python
def change_postion(X, Y, beta, std, alpha_E = 2, alpha_L =1):
    position_enter = []
    for i in range(len(X)):
        if abs(Y[i]-beta*X[i]) > alpha_E *std and len(position_enter)%2 == 0:
            position_enter.append(i)
        elif abs(Y[i]-beta*X[i]) < alpha_L *std and len(position_enter)%2 == 1:
            position_enter.append(i)

    return position_enter
```

```python
def plot_spread(s1, s2):
    model = LinearRegression(fit_intercept = False).fit(sp_data[s1].values.reshape(-1, 1), sp_data[s2].values.reshape(-1, 1))
    beta = model.coef_[0][0]
    std = np.std((sp_data[s2].values - sp_data[s1].values * beta))
    
    plt.plot((sp_data[s2].values - sp_data[s1].values * beta)/std, color = 'royalblue', label = 'Spread')
    plt.axhline(2, color = 'firebrick', label = 'Enter Postion')
    plt.axhline(-2,color = 'firebrick')
    
    plt.axhline(1, color = 'firebrick', linestyle = ':', label = 'Leave Position')
    plt.axhline(-1,color = 'firebrick', linestyle = ':')
    plt.axhline(0, color = 'k', linestyle = '--')
    plt.ylabel(r"$(Y_t - \beta X_t)/\sigma_u$")
    plt.xlabel("Day")
    plt.title(f'{s2} - {s1} Spread')
    plt.legend()

    change_positions = change_postion(sp_data[s1].values, sp_data[s2].values, beta, std)
    if change_positions is not None:    
        for ind in change_positions:
            y = (sp_data[s2].values[ind] - sp_data[s1].values[ind] * beta)/std
            plt.scatter(ind, y, color = 'green' if abs(y) > 2 else 'firebrick')
    
    plt.show()

plot_spread('LW', 'RSG')
plot_spread('BG', 'DD')
plot_spread('ADM', 'AVY')
plot_spread('HON', 'WHR')
plot_spread('ADM', 'DOW')
```

Whenever the solid line is crossed to the top (say day 40ish) we will take out a short position (Green dot) until the dotted line is crossed on day 50ish (Red dot). The opposite ias true when we cross the bottom red line. 

We now need to figure out how much money we make on each trade. For this let us assume that we want to put $500 into the position, this results in

- We short sell \$1000 worth of stock
- With this \$1000 gain, we buy long the other stock in the pair
- We taken the long stocks and the \$500 cash and put it in a margin account, the value of which must be 150% of the shorted stock.

i.e. we end up with a position long and short twice the value that we put in. 

```python
def n_stock_2_buy(position, price_X, price_Y):
    return (2*position)//price_X, (2*position)//price_Y

def price_pair_position(n_X, n_Y, price_X, price_Y): # Give the gain in opening a position (note when we close we get the negative)
    return -(n_X*price_X + n_Y*price_Y)
```

```python
@dataclass
class Position():
    ID : str # A identifier for the pairs that we have traded
    start_date : float # When we open the position
    amount : float # The amount that we want to put into the position
    stocks : list[tuple[str, int]] # A list of tuple where the first part is the stock and the second is the number of stocks
    end_date : float | None = None # End date, will be set once the position is closed
    

    def __len__(self):
        if self.end_date is None:
            print("Position must be closed")
            return 
        return self.end_date - self.start_date
    
    def is_open(self) -> bool:
        """Returns true if the position is yet to be closed """
        return self.end_date is None

    def close_position(self, end_date : float) -> None:
        self.end_date = end_date

    def returns(self, stock_prices : list) -> float:
        if self.end_date is None:
            print("Position must be closed to calculate profit") # CHANGE THIS TO AN ERROR MESSAGE
        return (1/(self.amount * len(self))) * (1+0*sum((stock_prices[i][self.end_date]*self.stocks[i][1] for i in range(len(stock_prices)))) - 0*sum((stock_prices[i][self.start_date]*self.stocks[i][1] for i in range(len(stock_prices)))))
    

    def annualised_returns(self, stock_prices) -> float:
        return self.returns(stock_prices) * (252/len(self))
```

```python
def calculate_returns(s1, s2):
    model = LinearRegression(fit_intercept = False).fit(sp_data[s1].values.reshape(-1, 1), sp_data[s2].values.reshape(-1, 1))
    beta = model.coef_[0][0]
    std = np.std((sp_data[s2].values - sp_data[s1].values * beta))
    
    change_position = change_postion(sp_data[s1].values, sp_data[s2].values, beta, std, alpha_E = 2, alpha_L =1)
    returns, positions = [], []
     
    for enter, leave in zip(change_position[::2],change_position[1::2]):    
        nX, nY = n_stock_2_buy(500,sp_data[s1].values[enter], sp_data[s2].values[enter])
        nX, nY = (nX, -nY) if (sp_data[s2].values[enter] - sp_data[s1].values[enter] * beta) > 0 else (-nX, nY)
        
        positions.append(Position(f"{s1}:{s2}", enter, 500, [(s1, nX), (s2, nY)], leave).returns([sp_data[s1].values, sp_data[s2].values])) 
        
        returns.append(price_pair_position(nX, nY, sp_data[s1].values[enter], sp_data[s2].values[enter]) - 
              price_pair_position(nX, nY, sp_data[s1].values[leave], sp_data[s2].values[leave]))
    
    
    return positions

returns = []
print(calculate_returns('LW', 'RSG'))
print(calculate_returns('BG', 'DD'))
print(calculate_returns('ADM', 'AVY'))
print(calculate_returns('HON', 'WHR'))
print(calculate_returns('ADM', 'DOW'))

returns.extend(calculate_returns('LW', 'RSG'))
returns.extend(calculate_returns('BG', 'DD'))
returns.extend(calculate_returns('ADM', 'AVY'))
returns.extend(calculate_returns('HON', 'WHR'))
returns.extend(calculate_returns('ADM', 'DOW'))

print((np.mean(returns)), np.std(returns))
```

## Saving Stocks Beta and Standard Deviation

```python
with open('trading_pairs.csv', 'r', newline='') as file:
    # Create a CSV reader object
    reader = csv.reader(file)
    pairs = [line for line in reader]
```

```python
with open('trading_pairs_info.csv', 'w', newline='') as file:
    # Create a CSV writer object
    writer = csv.writer(file)
    
    # Write each row of data to the CSV file
    for pair in pairs:
        model = LinearRegression(fit_intercept = False).fit(sp_data[pair[0]].values.reshape(-1, 1), sp_data[pair[1]].values.reshape(-1, 1))
        beta = model.coef_[0][0]
        std = np.std((sp_data[pair[1]].values - sp_data[pair[0]].values * beta))
        
        writer.writerow([pair[0], pair[1], beta, std])
```
